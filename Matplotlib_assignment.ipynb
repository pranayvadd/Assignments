```python
# TASK 2.7

plt.figure(figsize=(12,8))

x  = sales.month_number

y1 = sales.facecream.sum()
y2 = sales.facewash.sum()
y3 = sales.toothpaste.sum()
y4 = sales.bathingsoap.sum()
y5 = sales.shampoo.sum()
y6 = sales.moisturizer.sum()

data = [y1, y2, y3, y4, y5, y6]

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']

plt.axis("equal")

plt.pie(data, labels=labels, autopct="%1.1f%%")

plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

```python
# TASK 2.8

plt.figure(figsize=(12,6))

x  = sales.month_number
y2 = sales.facewash
y4 = sales.bathingsoap

plt.subplot(2,1,1)
plt.plot(x, y4, label = 'Bathing soap Sales Data', marker='o', color='k', linewidth=3)
plt.title('Sales data of bathing soap')

plt.subplot(2,1,2)
plt.plot(x, y2, label = 'Face wash Sales Data', marker='o', color='r', linewidth=3)
plt.title('Sales data of face wash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(x)
#plt.yticks([1500, 2000, 7500, 10000, 12500, 15000])
plt.show()
```

```python
# TASK 2.9

plt.figure(figsize=(12,8))

x  = sales.month_number
y1 = sales.facecream
y2 = sales.facewash
y3 = sales.toothpaste
y4 = sales.bathingsoap
y5 = sales.shampoo
y6 = sales.moisturizer

plt.plot([], [], label = 'Face cream', color='m', linewidth=3)
plt.plot([], [], label = 'Face wash', color='c', linewidth=3)
plt.plot([], [], label = 'Tooth paste', color='r', linewidth=3)
plt.plot([], [], label = 'Bathing soap', color='k', linewidth=3)
plt.plot([], [], label = 'Shampoo', color='g', linewidth=3)
plt.plot([], [], label = 'Moisturizer', color='y', linewidth=3)

plt.stackplot(x, y1, y2, y3, y4, y5, y6, colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('All products sales data using stock plat')
plt.show()
```
