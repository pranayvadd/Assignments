Create a given matrix, and also remove all the nan values from that matrix.

image.png

```python
x = np.array([[10.5, 22.5,np.nan],[41.0, 52.5, np.nan]])
print(x)

print("\nRemove all columns containing non-numeric elements ")
print(x[:, ~np.isnan(x).any(axis=0)])
```

Make a percentile function that will take two arguments,first argument is array and second argument as percentile and function will print their percentile value, as shown in the below image.

###Example:

image.png

```python
a = np.array([20, 2, 7, 1, 34])
print(a)

p50 = np.percentile(a, 50) # return 50th percentile
print ('50th percentile of a :',p50)

p25 = np.percentile(a, 25) # return 25th percentile
print ('25th percentile of a :',p25)

p75 = np.percentile(a, 75) # return 75th percentile
print ('75th percentile of a :',p75)
```
